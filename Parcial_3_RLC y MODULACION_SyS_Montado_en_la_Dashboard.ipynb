```python
%%writefile app.py

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import cheby2, filtfilt, freqz, TransferFunction, step, bode, impulse, tf2zpk
from yt_dlp import YoutubeDL
import soundfile as sf
import os
from matplotlib.patches import Circle

# Configuración inicial de la página
st.set_page_config(page_title="Parcial 3 - Señales y Sistemas", page_icon="📊", layout="wide")

# Barra lateral para seleccionar el punto del parcial y mostrar información general
with st.sidebar:
    st.image("https://lh4.googleusercontent.com/proxy/WNtyuTbDjnnITJFxg1dlI63L0jfIMRf0CIKg75VavFd3ameUuokpEiXIZvafO0UbA3rGKkhjDZ2HFtRWcGiPIn7Syd37PqnCrQuXFNHguRRPYm__safRJi9Q", caption="", width=250)
    st.title("Parcial 3 - Señales y Sistemas")
    st.subheader("Universidad Nacional de Colombia - Sede Manizales")
    st.write("**Integrantes:**")
    st.write("- Andrés Fernando Sosapanta López")
    st.write("- Juan Camilo Perdomo Soto")

    # Menú desplegable para seleccionar el punto del parcial
    punto = st.selectbox("Selecciona el punto del parcial",
                         ("Punto 1: Simulación Masa-Resorte-Amortiguador",
                          "Punto 2: Modulación AM con Filtro Pasa Bajas (Chebyshev II)"))

# Función para descargar y convertir el audio de YouTube
def download_audio_from_youtube(url, output_file='downloaded_song.wav'):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'wav',
            'preferredquality': '192',
        }],
        'outtmpl': 'temp_audio.%(ext)s',
    }
    with YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])
    # Convertir el archivo a wav
    os.system(f"ffmpeg -i temp_audio.wav -ar 48000 -ac 1 {output_file}")
    os.remove('temp_audio.wav')
    return output_file

# Función para simular el sistema masa-resorte-amortiguador
def simulate_system(zeta, c, k):
    m = (c**2) / (4 * (zeta**2) * k) if zeta > 0 else 1  # Cálculo de masa
    wn = np.sqrt(k / m)  # Frecuencia natural
    system = TransferFunction([wn**2], [1, 2 * zeta * wn, wn**2])

    # Respuesta al escalón
    t, y_step = step(system)
    fig, ax = plt.subplots()
    ax.plot(t, y_step)
    ax.set_title('Respuesta al Escalón del Sistema')
    ax.set_xlabel('Tiempo [s]')
    ax.set_ylabel('Desplazamiento [y(t)]')
    st.pyplot(fig)

    # Respuesta al impulso
    t, y_impulse = impulse(system)
    fig, ax = plt.subplots()
    ax.plot(t, y_impulse)
    ax.set_title('Respuesta al Impulso del Sistema')
    ax.set_xlabel('Tiempo [s]')
    ax.set_ylabel('Desplazamiento [y(t)]')
    st.pyplot(fig)

    # Diagrama de Bode
    w, mag, phase = bode(system)
    fig, (ax1, ax2) = plt.subplots(2, 1)
    ax1.semilogx(w, mag)
    ax1.set_title('Diagrama de Bode - Magnitud')
    ax1.set_ylabel('Magnitud [dB]')
    ax1.grid()

    ax2.semilogx(w, phase)
    ax2.set_title('Diagrama de Bode - Fase')
    ax2.set_xlabel('Frecuencia [rad/s]')
    ax2.set_ylabel('Fase [grados]')
    ax2.grid()
    st.pyplot(fig)

    # Diagrama de polos y ceros
    z, p, k = tf2zpk([wn**2], [1, 2 * zeta * wn, wn**2])
    fig, ax = plt.subplots()
    ax.scatter(np.real(z), np.imag(z), s=50, marker='o', label='Ceros')
    ax.scatter(np.real(p), np.imag(p), s=50, marker='x', label='Polos', color='red')
    unit_circle = Circle((0, 0), radius=1, fill=False, color='black', ls='solid', alpha=0.9)
    ax.add_patch(unit_circle)
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_title('Diagrama de Polos y Ceros')
    ax.grid(True)
    ax.set_aspect('equal', 'box')
    st.pyplot(fig)

# Función para simular la modulación AM y filtro pasa bajas Chebyshev II
def simulate_am_modulation(song, sample_rate):
    ti, tf = 20, 25  # Segundos a procesar
    m = song[int(ti*sample_rate):int(tf*sample_rate)]
    tt = np.arange(ti, tf, 1/sample_rate)

    # Mostrar y reproducir señal mensaje (audio original)
    st.write("### Señal Mensaje (Audio Original)")
    st.audio(m, format='audio/wav', sample_rate=sample_rate)

    fig, ax = plt.subplots()
    ax.plot(tt, m)
    ax.set_xlabel('$tiempo [s]$')
    ax.set_ylabel('$m(t)$')
    st.pyplot(fig)

    # Fourier señal mensaje
    Mw = np.fft.rfft(m)
    vf = np.fft.rfftfreq(m.shape[0], 1/sample_rate)
    fig, ax = plt.subplots()
    ax.plot(vf, np.abs(Mw))
    ax.set_title('Espectro señal mensaje')
    ax.set_xlabel(r'$f[Hz]$')
    ax.set_ylabel(r'$|M[f]|$')
    st.pyplot(fig)

    # Modulación AM
    Fo = 15000  # Frecuencia portadora
    c = np.cos(2*np.pi*Fo*tt)

    # Mostrar y reproducir portadora
    st.write("### Señal Portadora")
    st.audio(c, format='audio/wav', sample_rate=sample_rate)

    fig, ax = plt.subplots()
    ax.plot(tt, c)
    ax.set_xlabel('$tiempo [s]$')
    ax.set_ylabel('$c(t)$')
    st.pyplot(fig)

    # Señal modulada
    y = 2 * m * c  # Señal modulada

    # Mostrar y reproducir señal modulada
    st.write("### Señal Modulada (AM)")
    st.audio(y, format='audio/wav', sample_rate=sample_rate)

    fig, ax = plt.subplots()
    ax.plot(tt, y)
    ax.set_xlabel('$tiempo [s]$')
    ax.set_ylabel('$y(t)$')
    st.pyplot(fig)

    # Demodulación
    d = y * c  # Señal demodulada

    # Mostrar y reproducir señal demodulada
    st.write("### Señal Demodulada")
    st.audio(d, format='audio/wav', sample_rate=sample_rate)

    fig, ax = plt.subplots()
    ax.plot(tt, d)
    ax.set_xlabel('$tiempo [s]$')
    ax.set_ylabel('$d(t)$')
    st.pyplot(fig)

    # Filtro pasa bajas Chebyshev tipo II
    N = 9  # Aumentar el orden del filtro
    ripple = 10  # Reducir ripple
    Wn = 5000 / (sample_rate / 2)  # Reducir la frecuencia de corte a 5 kHz
    b, a = cheby2(N, ripple, Wn, btype='low', output='ba')

    # Filtrar señal demodulada
    xf = filtfilt(b, a, d)
    me = (2 / 2) * xf  # Recuperar señal mensaje

    # Guardar temporalmente la señal filtrada
    sf.write('filtered_audio_cheby.wav', me, sample_rate)

    # Mostrar y reproducir señal recuperada
    st.write("### Señal Recuperada (Audio Filtrado - Chebyshev II)")
    st.audio('filtered_audio_cheby.wav', format='audio/wav')

    fig, ax = plt.subplots()
    ax.plot(tt, me)
    ax.set_xlabel('$tiempo [s]$')
    ax.set_ylabel('$m_e(t)$')
    st.pyplot(fig)

    # Fourier señal recuperada
    Mew = np.fft.rfft(me)
    fig, ax = plt.subplots()
    ax.plot(vf, np.abs(Mew))
    ax.set_title('Espectro señal recuperada')
    ax.set_xlabel(r'$f[Hz]$')
    ax.set_ylabel(r'$|M_e[f]|$')
    st.pyplot(fig)

   # Diagrama de Bode (Magnitud y Fase)
    w, h = freqz(b, a, fs=sample_rate)

    # Magnitud
    fig, ax = plt.subplots()
    ax.semilogx(w, 20 * np.log10(abs(h)))
    ax.set_title('Diagrama de Bode - Magnitud')
    ax.set_xlabel('Frecuencia [Hz]')
    ax.set_ylabel('Magnitud [dB]')
    ax.grid()
    st.pyplot(fig)

    # Fase
    fig, ax = plt.subplots()
    ax.semilogx(w, np.angle(h))
    ax.set_title('Diagrama de Bode - Fase')
    ax.set_xlabel('Frecuencia [Hz]')
    ax.set_ylabel('Fase [radianes]')
    ax.grid()
    st.pyplot(fig)

    # Gráfico de ceros y polos
    z, p, k = tf2zpk(b, a)
    fig, ax = plt.subplots()
    ax.scatter(np.real(z), np.imag(z), s=50, marker='o', label='Ceros')
    ax.scatter(np.real(p), np.imag(p), s=50, marker='x', label='Polos', color='red')
    ax.axhline(0, color='black',linewidth=1)
    ax.axvline(0, color='black',linewidth=1)
    unit_circle = Circle((0, 0), radius=1, fill=False, color='black', ls='solid', alpha=0.9)
    ax.add_patch(unit_circle)
    ax.set_title('Diagrama de Polos y Ceros')
    ax.grid()
    ax.set_aspect('equal', 'box')
    st.pyplot(fig)

# Desarrollo del punto 1: Simulación
if punto == "Punto 1: Simulación Masa-Resorte-Amortiguador":
    st.write("### Simulación Sistema Masa-Resorte-Amortiguador")

    zeta = st.number_input("Ingresa el valor del factor de amortiguamiento ζ:", min_value=-5.0, max_value=5.0, value=0.5, step=0.1, key="zeta")
    c = st.number_input("Ingresa el valor de c (no puede ser 0):", min_value=0.1, value=1.0, step=0.1, key="c")
    k = st.number_input("Ingresa el valor de k (no puede ser 0):", min_value=0.1, value=1.0, step=0.1, key="k")

    if st.button("Simular", key="simulate_button"):
        simulate_system(zeta, c, k)

# Desarrollo del punto 2: Modulación AM con Filtro Pasa Bajas (Chebyshev II)
elif punto == "Punto 2: Modulación AM con Filtro Pasa Bajas (Chebyshev II)":
    st.write("### Modulación AM y Filtro Pasa Bajas Chebyshev II con canción de YouTube")

    youtube_url = st.text_input("Ingresa el enlace de YouTube de la canción")

    if youtube_url:
        with st.spinner("Descargando y procesando la canción..."):
            downloaded_song_path = download_audio_from_youtube(youtube_url)

        if downloaded_song_path:
            song, sample_rate = sf.read(downloaded_song_path)
            if len(song) >= 5 * sample_rate:  # Verificar que la longitud sea suficiente
                simulate_am_modulation(song, sample_rate)
                # Eliminar archivo temporal después de reproducir
                os.remove('filtered_audio_cheby.wav')
            else:
                st.error("La canción es demasiado corta. Sube una canción de al menos 5 segundos.")
```
